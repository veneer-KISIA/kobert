```python
import json

data_file = 'SXNE2202211218'
with open(f'./NIKL_NE_2022_v1.0/{data_file}.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
```

```python
labels = dict()

for document in data['document']:
    for sentence in document['sentence']:
            for ne in sentence['NE']:
                label = ne['label']
                word = ne['form']

                if label in labels.keys():
                     labels[label].update([word])
                else:
                     labels[label] = set()
                     labels[label].update([word])
```

```python
with open(f'./data/label_list_{data_file}.json', 'w', encoding='utf-8') as f:
    new_dict = {}
    for k, v in labels.items():
        new_dict[k] = list(v)

    json.dump(new_dict, f,indent=4, ensure_ascii=False)
```
